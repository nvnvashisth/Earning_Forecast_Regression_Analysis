```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("SP500_FUND_A_16052019.csv", index_col=0)
```

```python
df[:1]
```

```python
type(df.columns)
```

```python
data = df[['ticker','fyear','shr_y','ass_tot_y','goodwill_y','inc_rea_y','invest_y','rev_def_y','dps_y'
    ,'eps_y','exp_r_d_y','marg_profit_y','rev_grw_y','cf_inv_y','exp_cap_y'
    ,'ncf_bad_y','ncf_iad_y','g_cf_op_y','g_dps_y','g_eps_y','g_inv_y'
    ,'be_ps_y','fcf_ps_y','inc_qua_y','payout_y','turn_inv_y']]

data
```

##### It has null values "NaN", which is important to remove.

```python
data.isnull().any()
```

##### Describe help to provide overview of the data, which count, mean, stf , quantile range and min and max. This helps to take decision which columns need to be normalise or which column has bigger variance etc.

```python
data.describe()
```

##### Taking count of available null values. As we see there are wide range of null values which are present in the dataset. The entire dataset has 5k values and if we remove all of these null values, the dataset will remain with 130 column. So it is not wise to remove all, as we have very small dataset. In case, if we had 1M samples, then we could have thought of remove, which will not really effect the model, but in case of small dataset, we need to be very picky which columns is important to select and remove values from.

```python
data.isna().sum() 
```

##### Removing null values, those values create problem while scaling and could generate infinite values.

```python
data = data[np.isfinite(data['shr_y'])]
```

```python
data
```

###### Also we will also not remove other NaN values from entire dataset, as this could reduce the total number of samples. Hence we will convert all NaN values to 0, hence this will not contributed in modelling

```python
data = data.replace(np.NaN, 0)
```

```python
data
```

##### We have really high values of many column 10^8, hence it is important to scale down all values. Therefore, we decided to divided entire dataset except 'fyear' and 'ticker' with 'shr_y'. This will bring all values between computable range.

```python
scaled_data = data.iloc[:,3:].div(data.shr_y, axis=0)
```

```python
scaled_data
```

```python
data_temp = data[['ticker','fyear','shr_y']]
```

```python
data_temp
```

##### Scaled dataset

```python
data = pd.concat([data_temp, scaled_data],axis=1)
```

```python
data
```

```python
data_vars = data.columns.values.tolist()
```

##### As we can see the dataset contain lot of outlier after plotting into this boxplot. We plot for just 'shr_y' and then 'fyear' vs 'shr_y'. It is often assumed and know that, the most of the outlier usually lies at 0-5 and 95-100 percentile range. Hence we can consider for removing that.

```python
import matplotlib.pyplot as plt
import seaborn as sn
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(20,10))
sn.boxplot(data=data[['shr_y']], ax=ax1)
ax1.set(title="Boxplot for shr_y variables")
sn.boxplot(x=data['fyear'],y=data['shr_y'], ax=ax2)
ax2.set(title="Boxplot for fyear vs shr_y")
```

```python
data.quantile(.95, axis = 0) 
```

```python
data.quantile(.05, axis = 0) 
```

```python
data = data[data.shr_y < data.shr_y.quantile(.95)]
```

##### We have plotted again after removing the outlier from 5 and 95 percentile. And it is visible that, the median has been shift and outlier are in 75 percentile range now.

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(20,10))
sn.boxplot(data=data[['shr_y']], ax=ax1)
ax1.set(title="Boxplot for shr_y variables")
sn.boxplot(x=data['fyear'],y=data['shr_y'], ax=ax2)
ax2.set(title="Boxplot for fyear vs shr_y")
```

##### Creating Y variable for the prediction, which Forward Earning per share ('feps_y'). This can be calculated as Book Value per Share / Earing per share (be_ps_y/eps_y).   

```python
data['feps_y'] = data['be_ps_y'].div(data.eps_y, axis=0)
```

```python
data['feps_y'].replace([np.inf, -np.inf], 0,inplace=True)
```

```python
max(data['feps_y'])
```

```python
import statsmodels.formula.api as smf
```

```python
#lm = smf.ols(formula='shr_y ~ ass_tot_y + goodwill_y + inc_rea_y + invest_y + rev_def_y + dps_y + eps_y + exp_r_d_y + marg_profit_y + rev_grw_y + cf_inv_y + exp_cap_y + ncf_bad_y + ncf_iad_y + g_cf_op_y + g_dps_y + g_eps_y + g_inv_y + be_ps_y + fcf_ps_y + inc_qua_y + payout_y + turn_inv_y', data=data).fit()
lm = smf.ols(formula='feps_y ~ ass_tot_y + goodwill_y + inc_rea_y + invest_y + rev_def_y + dps_y + eps_y + exp_r_d_y + marg_profit_y + rev_grw_y + cf_inv_y + exp_cap_y + ncf_bad_y + ncf_iad_y + g_cf_op_y + g_dps_y + g_eps_y + g_inv_y + be_ps_y + fcf_ps_y + inc_qua_y + payout_y + turn_inv_y', data=data).fit()
```

```python
lm.params
```

Good F-statistic value mean better model. Also for feature selection, we go with p < 0.05 and select only those feature in

```python
lm.summary()
```

##### Selecting only those features which has p value < 0.05. 

```python
#lm = smf.ols(formula='shr_y ~ ass_tot_y + goodwill_y + inc_rea_y + invest_y + rev_def_y + dps_y + eps_y + exp_r_d_y + rev_grw_y + exp_cap_y + be_ps_y + fcf_ps_y + payout_y', data=data).fit()
lm = smf.ols(formula='feps_y ~ goodwill_y + inc_rea_y + dps_y + eps_y + rev_grw_y + exp_cap_y + be_ps_y + fcf_ps_y + inc_qua_y + payout_y + turn_inv_y', data=data).fit()
```

```python
lm.summary()
```

##### Predicting y value

```python
#y_fit = lm.predict(data[['ass_tot_y', 'goodwill_y', 'inc_rea_y', 'invest_y', 'rev_def_y', 'dps_y', 'eps_y', 'exp_r_d_y', 'rev_grw_y','exp_cap_y' ,'be_ps_y' ,'fcf_ps_y','payout_y']])
y_fit = lm.predict(data[['goodwill_y','inc_rea_y','dps_y','eps_y','rev_grw_y','exp_cap_y','be_ps_y','fcf_ps_y','inc_qua_y','payout_y','turn_inv_y']])           
```

```python
y_fit
```

#### Calculating the mean square error. How much is it deviated from the original value

```python
resid = data['feps_y'] - y_fit
```

```python
rss = np.sum(resid**2)
```

```python
MSE = rss/(lm.nobs-2)
```

```python
MSE
```
